Import Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Load the Data

```python
# Load the data
df = pd.read_csv("C:\\Users\\ACER\\Downloads\\API_SP.POP.csv")

# Display the first few rows to understand the structure
df.head()
```

```python
print(df.columns)
```

**Data Preparation**

This is population data for the most recent year, 2023.

```python
# Filter the dataset for the most recent year (2023)
df_2023 = df[['Country Name', '2023']]

# Drop any rows where the population value is missing
df_2023 = df_2023.dropna()

# Display the filtered data
df_2023.head()
```

**Create a Bar Chart**

Visualize the population distribution across the top 10 most populated countries.

```python
# Sort the data by population in descending order and select the top 10 countries
top_countries = df_2023.sort_values(by='2023', ascending=False).head(10)

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(top_countries['Country Name'], top_countries['2023'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Population in 2023')
plt.title('Top 10 Most Populated Countries in 2023')
plt.xticks(rotation=45)
plt.show()
```

The chart shows global population groups, with "World" being the largest, followed by income-based and demographic classifications like "Low & middle income" and "IDA & IBRD total."
